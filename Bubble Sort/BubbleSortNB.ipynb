```python
import matplotlib.pyplot as plt
import numpy as np
import time
from IPython.display import clear_output
```

```python
def bubble_sort(arr, draw_func):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
            draw_func(arr, j, j+1)
```

```python
def draw_bars(arr, idx1=None, idx2=None):
    clear_output(wait=True)
    plt.figure(figsize=(10,4))
    bars = plt.bar(range(len(arr)), arr, color='skyblue')
    if idx1 is not None:
        bars[idx1].set_color('red')
    if idx2 is not None:
        bars[idx2].set_color('green')
    plt.title("Live Bubble Sort Animation")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.show()
```

```python
# Example usage:
np.random.seed(0)
data = np.random.randint(1, 100, 20)
draw_bars(data)
start_time = time.time()
bubble_sort(data, draw_bars)
end_time = time.time()
elapsed = end_time - start_time
print(f"Sorting completed in {elapsed:.2f} seconds.")
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
def InsertionSort(arr, draw_fun):
    n = len(arr)
    for i in range(1,n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
        draw_fun(arr, i, j)
```

```python
def draw_fun(arr, idx1 = None, idx2 = None):
    clear_output(wait=True)
    plt.clf()
    plt.bar(range(len(arr)), arr, color='blue')
    if idx1 is not None and idx2 is not None:
        plt.bar(idx1, arr[idx1], color='red')
        plt.bar(idx2, arr[idx2], color='green')
    plt.title('Insertion Sort Visualization')
    plt.xlabel('Index')
    plt.ylabel('Value')
```

```python
np.random.seed(611)
def generate_random_array(size, min_value=0, max_value=100):
    return np.random.randint(min_value, max_value, size).tolist()
start_time = time.time()
InsertionSort(generate_random_array(20), draw_fun)
end_time = time.time()
print(f"Time taken: {end_time - start_time:.6f} seconds")
```
